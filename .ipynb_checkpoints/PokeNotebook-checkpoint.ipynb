# Ujian Pokemon Recommendation

```python
import numpy as np 
import pandas as pd 
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity 
import requests
```

#### Import CSV

```python
poke = pd.read_csv("Pokemon.csv")
poke
```

```python
poke.isnull().sum()
```

Data Type 1, Generation, dan Legendary aman

#### Cek Rekomendasi

```python
from sklearn.feature_extraction.text import CountVectorizer
ext = CountVectorizer()
```

```python
def kombinasi(i):
    return str(i['Type 1'])+ '$' +str(i['Generation'])+'$'+str(i['Legendary'])
poke['Atribute']= poke.apply(kombinasi,axis=1)
poke['Name']= poke['Name'].apply(lambda i: i.lower())

poke
```

```python
cov = CountVectorizer(tokenizer=lambda poke: poke.split('$'))
pokeUlti = cov.fit_transform(poke['Atribute'])
skorPoke = cosine_similarity(pokeUlti)
skorPoke
```

```python
pokeSuka = "pikachu"
indexSuka = poke[poke["Name"] == pokeSuka].index[0]
indexSuka

pokeRekom = list(enumerate(skorPoke[indexSuka]))
pokeSamaSortir = sorted(pokeRekom, key= lambda x:x[1], reverse= True)
pokeSamaSortir[:10]
```

```python
Rekom = []
for i in pokeSamaSortir[:7]:
    if i[0] != indexSuka:
        Rekom.append(poke.iloc[i[0]])
    
dfRekom = pd.DataFrame(Rekom)
dfRekom
```
